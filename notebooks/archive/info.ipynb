```python
import pandas as pd
import os

# Указываем путь к директории
directory_path = "data" + os.sep + "unpacked"
all_final_report = os.listdir(directory_path)

all_final_report
```

```python
df = pd.read_csv(
    "data/unpacked/thintergen_share_geno_VM2v2_1_part_3_FinalReport.txt.gz",
    sep="\t",
    header=9,
    compression="gzip",
    engine="pyarrow",
)
df
```

```python
lst_sample_id: list[set] = []

for i in all_final_report:
    df = pd.read_csv(
        f"{directory_path}{os.sep}{i}",
        sep="\t",
        header=9,
        compression="gzip",
        engine="pyarrow",
    )
    lst_sample_id.append(set(df["Sample ID"]))
```

```python
# Находим пересечения по Sample ID
from itertools import combinations


pairwise_intersections = {}

for (i, set1), (j, set2) in combinations(enumerate(lst_sample_id), 2):
    intersection_result = set1.intersection(set2)
    pairwise_intersections[(i, j)] = intersection_result
```

```python
print("Пересечения по Sample ID:")
count = 0
for i in pairwise_intersections.values():
    if i: 
        print(i)
    else:
        count += 1

if count == len(pairwise_intersections.values()):
    print("Пересечений нет. В каждой таблице уникальные организмы. Значит таблицы можно обрабывать по отдельности")
```
