```python
import pandas as pd
import matplotlib.pyplot as plt

# Загружаем результаты (PLINK разделяет данные пробелами)
df = pd.read_csv('/scratch/storageA/zaleski_bulls/bulls-vcf-pipeline/results/vcf/gwas_results.assoc.linear', sep='\s+')

# Отфильтруем только строки с тестом ADD (основной эффект аллеля)
results = df[df['TEST'] == 'ADD']

# Посмотрим на топ-10 самых крутых SNP (с самым маленьким P)
top_snps = results.sort_values('P').head(10)
print(top_snps)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = df[df['TEST'] == 'ADD'].dropna(subset=['P'])

# Считаем -log10(P)
df['-log10P'] = -np.log10(df['P'])

# Сортируем по хромосомам и позициям
df = df.sort_values(['CHR', 'BP'])
df['ind'] = range(len(df))

# Рисуем Manhattan Plot
fig = plt.figure(figsize=(15, 7))
ax = fig.add_subplot(111)
colors = ['#272e38', '#8d99ae']
x_labels = []
x_labels_pos = []

for i, (name, group) in enumerate(df.groupby('CHR')):
    group.plot(kind='scatter', x='ind', y='-log10P', color=colors[i % 2], ax=ax, s=15)
    x_labels.append(name)
    x_labels_pos.append((group['ind'].iloc[-1] + (group['ind'].iloc[0])) / 2)

# Линии значимости
ax.axhline(-np.log10(5e-8), color='red', linestyle='--', label='Genome-wide significance')
ax.axhline(-np.log10(1e-5), color='blue', linestyle=':', label='Suggestive evidence')

ax.set_xticks(x_labels_pos)
ax.set_xticklabels(x_labels)
ax.set_ylabel('-log10(P)')
plt.legend()
plt.title('Manhattan Plot for Milk Yield')
plt.savefig('manhattan.png', dpi=300)
print("График сохранен в manhattan.png")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# 1. Загрузка данных
df = pd.read_csv('/scratch/storageA/zaleski_bulls/bulls-vcf-pipeline/results/vcf/gwas_results.assoc.linear', sep='\s+')
df = df[df['TEST'] == 'ADD'].dropna(subset=['P'])

# 2. Подготовка p-values
p_values = np.sort(df['P'])
expected = np.arange(1, len(p_values) + 1) / (len(p_values) + 1)

# Переводим в -log10 масштаб
observed_log = -np.log10(p_values)
expected_log = -np.log10(expected)

# 3. Визуализация
plt.figure(figsize=(6, 6))
plt.scatter(expected_log, observed_log, c='blue', s=10, alpha=0.5)

# Рисуем идеальную диагональ (y = x)
max_val = max(max(expected_log), max(observed_log))
plt.plot([0, max_val], [0, max_val], color='red', linestyle='--')

plt.xlabel('Expected -log10(P)')
plt.ylabel('Observed -log10(P)')
plt.title('Q-Q Plot: Protein Yield GWAS')
plt.grid(alpha=0.3)

plt.savefig('qq_plot.png', dpi=300)
plt.show()
```
