```python
def create_TopGenomicSeq(row):
    first = row["TopGenomicSeq"].split("/")[0][-1]
    second = row["TopGenomicSeq"].split("/")[1][0]
    return f"[{first}/{second}]"


def create_SourceSeq_SNP(row):
    first = row["SourceSeq"].split("/")[0][-1]
    second = row["SourceSeq"].split("/")[1][0]
    return f"[{first}/{second}]"


bovinehd_manifest_df["TopGenomicSeq_SNP"] = bovinehd_manifest_df.apply(
    create_TopGenomicSeq, axis=1
)

bovinehd_manifest_df["SourceSeq_SNP"] = bovinehd_manifest_df.apply(
    create_SourceSeq_SNP, axis=1
)

display(bovinehd_manifest_df[bovinehd_manifest_df["Name"] == "ARS-BFGL-NGS-65067"])
```

```python
# def get_alt_from_normalized(row):
#     ref_allele = row["REF"]
#     # normalized_alleles = row['Normalized_SNP']
#     TopGenomicSeq_SNP = row["TopGenomicSeq_SNP"].strip("[]").split("/")
#     SourceSeq_SNP = row["SourceSeq_SNP"].strip("[]").split("/")
#     TopGenomicSeq_Left = row["TopGenomicSeq"].split("[")[0][-20:].upper()
#     SourceSeq_Left = row["SourceSeq"].split("[")[0][-20:].upper()


#     if int(row["MapInfo"]) == 0 or int(row["MapInfo"]) == 0:
#         return "DATA_ERROR"
#     base = tools.get_ref_base(
#         int(row["Chr_y"]), int(row["MapInfo"] - 1), int(row["MapInfo"] - 20)
#     )

#     if len(TopGenomicSeq_Left) != len(base) or len(SourceSeq_Left) != len(base):
#         print("base", len(base))
#         print("TopGenomicSeq_Left", len(TopGenomicSeq_Left))
#         print("SourceSeq_SNP_Left", len(SourceSeq_Left))

#     if base == TopGenomicSeq_Left and ref_allele in TopGenomicSeq_SNP:
#         return (
#             TopGenomicSeq_SNP[0]
#             if TopGenomicSeq_SNP[1] == ref_allele
#             else TopGenomicSeq_SNP[1]
#         )
#     elif base == SourceSeq_Left and ref_allele in SourceSeq_SNP:
#         return SourceSeq_SNP[0] if SourceSeq_SNP[1] == ref_allele else SourceSeq_SNP[1]
#     else:
#         return "DATA_ERROR"


def get_alt_from_normalized(row):
    complementarity = {"A": "T", "T": "A", "C": "G", "G": "C"}
    if pd.isnull(row["REF"]) or pd.isnull(row["SNP"]) or pd.isnull(row["RefStrand"]):
        return "DATA_ERROR"

    ref_allele = row["REF"]
    snp = row["SNP"].strip("[]").split("/")
    ref_strand = row["RefStrand"]

    if ref_strand == "+":
        if ref_allele in snp:
            return snp[0] if snp[1] == ref_allele else snp[1]
        else:
            return "DATA_ERROR"
    elif ref_strand == "-":
        # complement_ref = complementarity.get(ref_allele)
        snp = [complementarity[snp[0]], complementarity[snp[1]]]
        if ref_allele in snp:
            return snp[0] if snp[1] == ref_allele else snp[1]
            # ALT для - цепи
            # alt_on_minus_strand = snp[0] if snp[1] == complement_ref else snp[1]
            # # ALT Для + цепи
            # return complementarity.get(alt_on_minus_strand)
        else:
            return "DATA_ERROR"
    else:
        return "DATA_ERROR"


rdf["ALT"] = rdf.apply(get_alt_from_normalized, axis=1)


rdf
```
