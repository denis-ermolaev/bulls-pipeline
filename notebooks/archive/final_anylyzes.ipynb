```python
def create_is_correct(row, colummn_gen):
    # print(row["HORUS003910309615_x"])
    source_gen = row[f"{colummn_gen}_y"].split(":")[0]
    if str(row["INFO_x"][-3:]) != "IMP":
        return None
    if source_gen == "./.":
        return None
    imputed_non_phased = row[f"{colummn_gen}_x"][0:3].split("|")
    imputed_non_phased = "/".join(list(sorted(imputed_non_phased)))
    if imputed_non_phased == source_gen:
        return True
    else:
        return False


all = 0
correct_imput = 0

for i in ["3886_1_bwa.bam.sorted.bam", "3842_1_bwa.bam.sorted.bam"]:
    is_correct = usa_merge_for_correct.apply(lambda row: create_is_correct(row, i), axis = 1)
    all += len(is_correct[is_correct == True]) + len(is_correct[is_correct == False])
    correct_imput += len(is_correct[is_correct == True])
print(f"Импутация корректно восстановила {round((correct_imput/all)*100)}%")
print(f"{all=}, {correct_imput=}")
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import pearsonr

def gt_to_dosage(gt):
    """Преобразует строку генотипа в число альтернативных аллелей (0,1,2)"""
    if pd.isna(gt) or gt in ('./.', '.|.'):
        return np.nan
    # заменяем возможный разделитель '|' на '/' (если есть фазировка)
    gt = gt.replace('|', '/')
    alleles = gt.split('/')
    # сумма чисел даст количество альтернативных аллелей
    if not isinstance(int(alleles[0]) + int(alleles[1]), int):
        print(int(alleles[0]) + int(alleles[1]))
        print(type(int(alleles[0]) + int(alleles[1])))
    return int(alleles[0]) + int(alleles[1])

# Применяем ко всем колонкам с образцами (предположим, они все, кроме первой колонки 'SNP')
# Например, если образцы называются Cow1, Cow2, ...
# sample_cols = [col for col in df_rez_imputation_01.columns if col.startswith('HORUS')]  # подбери под свои названия
sample_cols = df_rez_imputation_01.columns[9:]
for col in sample_cols:
    df_rez_imputation_01[col] = df_rez_imputation_01[col].apply(gt_to_dosage)

# 1. Случайно перемешай список животных и раздели
animals = df_rez_imputation_01.columns[9:].tolist()
np.random.seed(123)  # для воспроизводимости
np.random.shuffle(animals)
half = len(animals) // 2
group1 = animals[:half]
group2 = animals[half:]

# 2. Функция для расчёта MAF по группе
def calc_maf(df, group):
    # Сумма альтернативных аллелей по группе
    alt_sum = df[group].sum(axis=1, skipna=True)
    # Общее число хромосом в группе (2 * число животных в группе)
    n_chrom = 2 * len(group)
    maf = alt_sum / n_chrom
    # MAF — это минорная частота, поэтому если >0.5, берём 1-maf
    maf = np.where(maf > 0.5, 1 - maf, maf)
    return maf

maf1 = calc_maf(df_rez_imputation_01, group1)
maf2 = calc_maf(df_rez_imputation_01, group2)

# Убираем SNP, где MAF=0 в обеих группах (они неинформативны)
mask = (maf1 > 0) & (maf2 > 0)
maf1_f = maf1[mask]
maf2_f = maf2[mask]

# Корреляция
r, p = pearsonr(maf1_f, maf2_f)
print(f'Корреляция между двумя половинами: r = {r:.4f}, p = {p:.2e}')

# График
plt.figure(figsize=(6,6))
plt.scatter(maf1_f, maf2_f, s=1, alpha=0.3)
plt.plot([0,0.5], [0,0.5], 'r--')
plt.xlabel('MAF группа 1')
plt.ylabel('MAF группа 2')
plt.title(f'Сравнение MAF в двух половинах выборки\nr = {r:.3f}')
plt.show()
```

```python
# # Импорты
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Загрузка PCA
eigenvec = pd.read_csv('/scratch/storageA/zaleski_bulls/bulls-vcf-pipeline/bin/new/for_pca/bulls_pca.eigenvec', sep='\s+', header=None)
eigenvec.columns = ['FID', 'IID'] + [f'PC{i}' for i in range(1, 11)]

# (опционально) фильтр по русским коровам
# russian_ids = ... 
# russian_pca = eigenvec[eigenvec['IID'].isin(russian_ids)].copy()
# Если PCA сделано только на русских, то просто:
russian_pca = eigenvec.copy()

# Выбор PCs для кластеризации (первые три)
X = russian_pca[['PC1', 'PC2', 'PC3']].values

# Определение оптимального числа кластеров (попробуем от 2 до 5)
sil = []
for k in range(2, 6):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10)
    labels = kmeans.fit_predict(X)
    sil.append(silhouette_score(X, labels))
    print(f'k={k}, silhouette={sil[-1]:.3f}')

# Выбираем k с максимальным силуэтом (или по визуальной оценке)
best_k = np.argmax(sil) + 2
print(f'Выбрано k = {best_k}')

# Финальная кластеризация
kmeans = KMeans(n_clusters=best_k, random_state=42, n_init=10)
russian_pca['cluster'] = kmeans.fit_predict(X)

# Сохраняем
russian_pca[['IID', 'cluster']].to_csv('clusters.csv', index=False)

# Визуализация
plt.figure(figsize=(10,8))
for clust in range(best_k):
    subset = russian_pca[russian_pca['cluster'] == clust]
    plt.scatter(subset['PC1'], subset['PC2'], label=f'Cluster {clust}', alpha=0.6, s=10)
plt.xlabel('PC1')
plt.ylabel('PC2')
plt.legend()
plt.title('PCA clusters')
plt.grid(True, alpha=0.3)
plt.show()
```
